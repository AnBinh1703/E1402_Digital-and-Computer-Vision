# Drawing on Video

Often as we analyze video using techniques like object detection or facial recognition, we'll want to draw an image on the video, like a box around a face.

Let's learn the basics of drawing on video with OpenCV!

```python
import cv2

cap = cv2.VideoCapture(0)  # or video path
if not cap.isOpened():
    raise RuntimeError("Cannot open camera/video source")

while True:
    ret, frame = cap.read()
    if not ret or frame is None:
        print("No frame received. Stopping.")
        break

    # Example drawing
    x, y, w, h = 50, 50, 200, 150
    cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 0, 255), 4)

    cv2.imshow("frame", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

## Interactive Drawing on Video

We won't really use this as often, but let's go through it just in case!

```python
import cv2
import numpy as np
```

```python
import cv2
import os

# Input video
video_path = r'D:\UMEF\E1402_Digital and Computer Vision\session5\camera\student_capture.avi'
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    raise RuntimeError(f"Không mở được video: {video_path}")

# Output folder + writer
out_dir = r'D:\UMEF\E1402_Digital and Computer Vision\session5\camera\video_output'
os.makedirs(out_dir, exist_ok=True)

fps = cap.get(cv2.CAP_PROP_FPS)
if fps <= 0:
    fps = 25
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

out_path = os.path.join(out_dir, 'rectangle_output.avi')
writer = cv2.VideoWriter(out_path, cv2.VideoWriter_fourcc(*'XVID'), fps, (w, h))

# Mouse states
pt1 = (0, 0)
pt2 = (0, 0)
drawing = False
rect_done = False

def draw_rectangle_drag(event, x, y, flags, param):
    global pt1, pt2, drawing, rect_done
    if event == cv2.EVENT_LBUTTONDOWN:
        pt1 = (x, y)
        pt2 = (x, y)
        drawing = True
        rect_done = False
    elif event == cv2.EVENT_MOUSEMOVE and drawing:
        pt2 = (x, y)
    elif event == cv2.EVENT_LBUTTONUP and drawing:
        pt2 = (x, y)
        drawing = False
        rect_done = True

cv2.namedWindow("Test")
cv2.setMouseCallback("Test", draw_rectangle_drag)

while True:
    ret, frame = cap.read()
    if not ret or frame is None:
        break

    if drawing or rect_done:
        cv2.rectangle(frame, pt1, pt2, (0, 0, 255), 2)

    cv2.imshow("Test", frame)
    writer.write(frame)

    key = cv2.waitKey(max(1, int(1000 / fps))) & 0xFF
    if key == ord('c'):
        rect_done = False
        drawing = False
        pt1, pt2 = (0, 0), (0, 0)
    elif key == ord('q'):
        break

cap.release()
writer.release()
cv2.destroyAllWindows()
print("Saved:", out_path)
```
