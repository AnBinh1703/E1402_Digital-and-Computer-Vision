**TASK: Use PIL and matplotlib to read and display the ../DATA/00-puppy.jpg image.**

```python
from PIL import Image
import matplotlib.pyplot as plt

# Read the image
img = Image.open('00-puppy.jpg')

# Display the image
plt.figure(figsize=(8, 6))
plt.imshow(img)
plt.title('Puppy Image')
plt.axis('off')
plt.tight_layout()
plt.show()
```

**TASK: Convert the image to a NumPy Array**

```python
img_array = np.array(img)
print(f"Image shape (Height, Width, Channels): {img_array.shape}")
print(f"Image data type: {img_array.dtype}")
img_array
```

**FINAL TASK: Use slicing to set the RED and GREEN channels of the picture to 0, then use imshow() to show the isolated blue channel**

```python
# Create a copy of the image array
blue_channel = img_array.copy()

# Set RED (channel 0) and GREEN (channel 1) to 0, keeping only BLUE (channel 2)
blue_channel[:, :, 0] = 0  # RED channel
blue_channel[:, :, 1] = 0  # GREEN channel

# Display just the blue channel
plt.figure(figsize=(8, 6))
plt.imshow(blue_channel)
plt.title('Blue Channel Only')
plt.axis('off')
plt.tight_layout()
plt.show()
```
