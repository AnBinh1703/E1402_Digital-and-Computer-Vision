## Project Task

**You only have one task here. Create a program that reads in a live stream from a camera on your computer (or if you don't have a camera, just open up a video file). Then whenever you click the left mouse button down, create a blue circle around where you've clicked. Check out the video for an example of what the final project should look like**

**Guide**

* Create a draw_circle function for the callback function
* Use two events cv2.EVENT_LBUTTONDOWN and cv2.EVENT_LBUTTONUP
* Use a boolean variable to keep track if the mouse has been clicked up and down based on the events above
* Use a tuple to keep track of the x and y where the mouse was clicked.
* You should be able to then draw a circle on the frame based on the x,y coordinates from the Event 

Check out the skeleton guide below:

```python
import cv2
import os

# ----- Input video paths (try AVI first, then MP4) -----
candidates = [
    r'D:\UMEF\E1402_Digital and Computer Vision\session5\camera\student_capture.avi',
    r'D:\UMEF\E1402_Digital and Computer Vision\session5\camera\student_capture.mp4',
]

cap = None
input_path = None
for p in candidates:
    c = cv2.VideoCapture(p)
    if c.isOpened():
        cap = c
        input_path = p
        break

if cap is None:
    raise RuntimeError("Cannot open input video (.avi/.mp4).")

# ----- Create output folder -----
output_dir = r'D:\UMEF\E1402_Digital and Computer Vision\session5\camera\video_output'
os.makedirs(output_dir, exist_ok=True)

# ----- Output writer -----
fps = cap.get(cv2.CAP_PROP_FPS)
if fps <= 0:
    fps = 25

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

output_path = os.path.join(output_dir, 'circle_output.avi')
writer = cv2.VideoWriter(
    output_path,
    cv2.VideoWriter_fourcc(*'XVID'),
    fps,
    (width, height)
)

if not writer.isOpened():
    cap.release()
    raise RuntimeError(f"Cannot create output video: {output_path}")

# ----- Mouse callback -----
center = (0, 0)
clicked = False

def draw_circle(event, x, y, flags, param):
    global center, clicked
    if event == cv2.EVENT_LBUTTONDOWN:
        center = (x, y)
        clicked = False
    elif event == cv2.EVENT_LBUTTONUP:
        center = (x, y)
        clicked = True

cv2.namedWindow('Test')
cv2.setMouseCallback('Test', draw_circle)

# ----- Process video -----
while True:
    ret, frame = cap.read()
    if not ret or frame is None:
        break

    if clicked:
        cv2.circle(frame, center, 40, (0, 0, 255), 2)

    cv2.imshow('Test', frame)
    writer.write(frame)  # save output video

    if cv2.waitKey(int(1000 / fps)) & 0xFF == ord('q'):
        break

cap.release()
writer.release()
cv2.destroyAllWindows()

print("Input :", input_path)
print("Saved :", output_path)
```
