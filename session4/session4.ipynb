```python
# Apply Sobel with different kernel sizes
sobel_results = {}
ksizes = [5, 7, 9]

for ksize in ksizes:
    # Apply Sobel X and Y with current ksize
    sx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=ksize)
    sy = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=ksize)
    
    # Convert to absolute values
    sx = cv2.convertScaleAbs(sx)
    sy = cv2.convertScaleAbs(sy)
    
    # Combine
    combined = cv2.addWeighted(sx, 0.5, sy, 0.5, 0)
    sobel_results[ksize] = combined

print("Sobel edge detection with different kernel sizes completed")
```
