# OpenCV with Video Files

Let's now open the recorded video from the last lecture, the file is called "video_capture.mp4", although you can use this code to open any major video format.

```python
# Run everything in one cell!
```

```python
import cv2
import os

save_path = r'D:\UMEF\E1402_Digital and Computer Vision\session5\camera\student_capture.avi'
os.makedirs(os.path.dirname(save_path), exist_ok=True)

cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise RuntimeError("Cannot open camera")

w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# AVI + XVID is usually most reliable in OpenCV on Windows
writer = cv2.VideoWriter(save_path, cv2.VideoWriter_fourcc(*'XVID'), 25, (w, h))
if not writer.isOpened():
    cap.release()
    raise RuntimeError("Cannot open VideoWriter")

try:
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        writer.write(frame)
        cv2.imshow("rec", frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
finally:
    cap.release()
    writer.release()
    cv2.destroyAllWindows()

print("Saved:", save_path)
```

```python
import cv2

video_path = r'D:\UMEF\E1402_Digital and Computer Vision\session5\camera\student_capture.avi'
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    raise RuntimeError(f"Cannot open: {video_path}")

fps = cap.get(cv2.CAP_PROP_FPS) or 25
delay = int(1000 / fps)

while True:
    ret, frame = cap.read()
    if not ret:
        break
    cv2.imshow("play", frame)
    if cv2.waitKey(delay) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
